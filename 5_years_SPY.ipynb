```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


spy_ticker = yf.Ticker("SPY")
spy_data = spy_ticker.history(period="5y")

print(spy_data.head(5))
```

```python
print(spy_data.isna().any().any())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(spy_data.index, spy_data['Close'], label='SPY Close Price')
plt.title('SPY 5-Year Historical Price Data')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
